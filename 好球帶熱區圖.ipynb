```python
data = pd.read_excel(excel_file_path, sheet_name=sheet_name)
```

```python
data
```

```python
col = ['umpireInChief','bName','bHand','pHand','strikes','side','inning','coordX','coordY','pType','call','result']
```

```python
df = data[col]
```

```python
df
```

```python
result_items = df['result'].unique()
result_items
```

```python
df = df[~df['result'].isin(['GO','2B', '1B', 'FO', 'SH', 'E', 'GIDP', 'FC','HBP', 'HR', '3B', 'IBB', 'SF', 'IGNORE', 'DP', 'D3S', 'ID', 'IH'])]
```

### FT:擦棒被捕
### CS:出棒過半

```python
call_items = df['call'].unique()
call_items
```

```python
df = df[~df['call'].isin(['F', 'SW', 'CS', 'FT_MISS', 'FOUL_BUNT','FT','TRY_BUNT', 'H', 'BUNT'])]
df= df.dropna(subset=['call'])
```

```python
pitch_items = df['pType'].unique()
pitch_items
```

```python
name_mapping = {
    'FF':'直球', 
    'CH':'變速球', 
    'FC':'卡特球', 
    'CU':'曲球', 
    'FO':'指叉球',
    'SL':'滑球', 
    'SI':'伸卡球',
    'KN':'彈指曲球', 
    'EP':'小便球'
}

# 新增一列"球員"，根据"Pitcher"列的值匹配中文名
df['球種'] = df['pType'].map(name_mapping)
```

```python
df
```

```python
umpire_items = df['umpireInChief'].unique()
umpire_items
```

```python
pitch_items = df['球種'].unique()
pitch_items
```

```python
umpire_name = "紀華文"
```

```python
# 設定篩選 1
#batter_name = "王威晨"
call = "S"
#strikes = 2
#inning = [7,8,9]
#pitchtype = ["卡特球"]
#side = "HOME"
pHand = "L"
bHand = ["R","L"]
filtered_df = df[(df['umpireInChief'] == umpire_name) 
                #& (df['bName'] == batter_name)
                & (df['call'] == call) 
                & (df['pHand'] == pHand)
                #& (df['side'] == side)
                #& (df['strikes'] == strikes)
                & df['bHand'].isin(bHand)
                #& df['inning'].isin(inning)
                #& df['球種'] .isin(pitchtype)
                ]
filtered_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 繪製散點圖
plt.figure(figsize=(8, 8))
plt.scatter(filtered_df['coordX'], filtered_df['coordY'], c='blue', alpha=0.6, edgecolors='w', s=100)

# 繪製好球帶
shadow_zone = plt.Polygon([[-66.67, 66.67], [66.67, 66.67], [66.67, -66.67], [-66.67, -66.67]], fill=None, edgecolor='orange', linewidth=2)
plt.gca().add_patch(shadow_zone)
strike_zone = plt.Polygon([[-50, 50], [50, 50], [50, -50], [-50, -50]], fill=None, edgecolor='red', linewidth=2)
plt.gca().add_patch(strike_zone)

# 設定圖表標題及軸標籤
plt.title(f'Umpire - Pitcher {pHand} vs Batter {bHand} - Strike Locations')
plt.xlabel('coordX')
plt.ylabel('coordY')
plt.xlim(-150, 150)
plt.ylim(-150, 150)
plt.grid(True)

# 顯示圖表
plt.show()
```

```python
# 繪製熱區圖
plt.figure(figsize=(8, 8))

# 使用 kdeplot 繪製核密度估計熱區圖
sns.kdeplot(
    x=filtered_df['coordX'], y=filtered_df['coordY'], 
    fill=True, cmap='Reds', thresh=0, levels=100
)

# 或者使用 histplot 繪製熱區圖
# sns.histplot(
#     x=filtered_df['coordX'], y=filtered_df['coordY'], 
#     bins=50, pmax=0.8, cmap='Blues'
# )

# 繪製好球帶
shadow_zone = plt.Polygon([[-66.67, 66.67], [66.67, 66.67], [66.67, -66.67], [-66.67, -66.67]], fill=None, edgecolor='orange', linewidth=2)
plt.gca().add_patch(shadow_zone)
strike_zone = plt.Polygon([[-50, 50], [50, 50], [50, -50], [-50, -50]], fill=None, edgecolor='black', linewidth=2)
plt.gca().add_patch(strike_zone)

# 設定圖表標題及軸標籤
plt.title(f'Umpire - Pitcher {pHand} vs Batter {bHand} - Strike Locations')
plt.xlabel('coordX')
plt.ylabel('coordY')
plt.xlim(-150, 150)
plt.ylim(-150, 150)
plt.grid(True)

# 顯示圖表
plt.show()
```
