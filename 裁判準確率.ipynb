```python
stadiums_of_interest = [
    '臺北市立天母棒球場 Tianmu Baseball Stadium',
    '臺中洲際棒球場 \tTaichung Intercontinental Baseball Stadium'
]

# 使用 loc 方法筛选数据
df2 = df[df['stadium'].isin(stadiums_of_interest)]

```

```python
def is_in_strike_zone(x, y):
    strike_zone_x_min = -50
    strike_zone_x_max = 50
    strike_zone_y_min = -50
    strike_zone_y_max = 50
    return strike_zone_x_min <= x <= strike_zone_x_max and strike_zone_y_min <= y <= strike_zone_y_max
```

```python
#用蒙特卡羅模擬 投球位置誤差
import numpy as np

def simulate_true_locations(coordX, coordY, num_simulations=700, x_std=0, y_std=0):
    simulated_locations = np.zeros((num_simulations, 2))
    for i in range(num_simulations):
        simulated_locations[i, 0] = np.random.normal(coordX, x_std)
        simulated_locations[i, 1] = np.random.normal(coordY, y_std)
    return simulated_locations
```

```python
#計算好球的機率
def calculate_strike_probability(simulated_locations):
    strikes = np.array([is_in_strike_zone(x, y) for x, y in simulated_locations])
    strike_probability = np.mean(strikes)
    return strike_probability
```

```python
def is_call_correct(row, strike_probability):
    call_is_strike = row['call'] == 'S'
    call_is_ball = row['call'] == 'B'
    
    # 判断為壞球
    if strike_probability < 0.15:
        return call_is_ball  # 如果可能是壞球，裁判叫了壞球，則正确
    
    # 判断好球情况
    if strike_probability > 0.85:
        return call_is_strike  # 如果可能是好球，裁判叫了好球，則正確    
    # 如果既不是很可能是好球也不是很可能是坏球，则不做判决
    return True
    
# 對於每一球，計算投球的真實位置
results = []
for index, row in df2.iterrows():
    simulated_locations = simulate_true_locations(row['coordX'], row['coordY'])
    strike_probability = calculate_strike_probability(simulated_locations)
    correct_call = is_call_correct(row, strike_probability)
    results.append(correct_call)

df2['correct_call'] = results
```

```python
summary = df2.groupby(['umpireInChief'])['correct_call'].value_counts().unstack().fillna(0)
summary = summary.reset_index(drop=False)
summary
```

```python
summary = pd.DataFrame(summary)
summary.columns = ['umpireInChief', 'False', 'True']
summary["準確率"]= summary["True"] / (summary["True"]+summary["False"])*100
summary
```

```python
False_total = summary['False'].sum()
True_total = summary['True'].sum()
```
